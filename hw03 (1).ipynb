### Q1.a

Recall that *sets* are collections where each value in the collection only appears once.

Using the `set` function, turn both of these strings into sets. Then use set methods to find

* How many letters are shared between the two names?
* Which letters are in "George Washington" that are not in "Abraham Lincoln"?
* How many letters are in neither president's name?


```python
gw = "George Washington".lower()
al = "Abraham Lincoln".lower()
az = "abcdefghijklmnopqrstuvwxyz"

# convert the string to sets of characters
gw_set = set(gw)
al_set = set(al)
az_set = set(az)

# find how many letters are in both sets
both = gw_set.intersection(al_set)

print("There are " + str(len(both)) + " characters shared between the two names")

# letters in "George Washington" but not in "Abraham Lincoln"
print(gw_set.difference(al_set))

# letters that are in neither set
neither = az_set.difference(gw_set.union(al_set))
print("There are " + str(len(neither)) + " characters that are in neither president's name")
```


### Q1.d

Write a function that takes a lunch order (a set of menu items) and returns the total price.

Write a function that takes a lunch order a returns true if the entire order is vegetarian.

Use list comprehensions to compute the price of each order and whether each order is vegtarian.

```python
# Hints: try running the following code
# ("this", "is", "a tuple")[1]
# [s.upper() for s in {"hello", "world"}]


def calculateTotalPrice(order):
    total = 0
    # use list comprehension to iterate over the order and sum the prices
    total = sum([menu[item][0] for item in order])
    return total

def isVegetarian(order):
    # use list comprehension to iterate over the order and check if all items are vegetarian
    return sum([menu[item][1] for item in order]) == len(order)

order = {"pasta", "coffee", "burger"}
print(order)
total_price = calculateTotalPrice(order)
is_vegetarian = isVegetarian(order)
print(total_price)
print(is_vegetarian)
```

Using **methods** for Series objects (i.e., methods called on the columns of the table. Compute the following:

* The name of the least expensive item on the list.
* The number of items that cost between 5 and 10 dollars (inclusive).
* The names all of vegetarian items that cost more than 10 dollars.

```python
# find the cheapest item on the list
min_index = menu_df["price"].argmin()
cheapest_item = menu_df["name"][min_index]
print("The cheapest item is " + cheapest_item)

# find the number of items that cost between 5 and 10 inclusive
price = menu_df["price"]
number_items = price[(price >= 5) & (price <= 10)].count()
print("There are " + str(number_items) + " items that cost between 5 and 10")

# I only want vegetarian true
vegetarian = menu_df["vegetarian"]
veg_items = menu_df[vegetarian]

# get the name of the vegetarian item that costs more than 10
is_more_than_10 = veg_items["price"] > 10
expensive_veg = veg_items[is_more_than_10]
print("The vegetarian item that costs more than 10 is " + expensive_veg["name"].values[0])
```

Let's use these techniques on some real data!

```python
spotify = pd.read_csv("https://github.com/UM-Data-Science-101/homework-03/raw/refs/heads/main/spotify.csv")
spotify.columns
```

Using Python and Pandas, answer the following questions:

* How many songs have a popularity greater than 80?
* For songs with a popularity greater than 80, what is the shortest song (in miliseconds)?
* What is the name of the song with popularity greater than 80 that has the slowest tempo?

```python
# number of popular songs
popularity = spotify["track.popularity"]
is_popular = popularity > 80
number = popularity[is_popular].count()
print("There are " + str(number) + " songs with popularity > 80")

# shortest song with popularity > 80
duration = spotify["duration_ms"]
shortest_song_index = duration[is_popular].argmin()

# get the name of the shortest song with the index
shortest_popular_song = spotify["track.name"][shortest_song_index]
print("The shortest song with popularity > 80 is: " + shortest_popular_song)

# slowest song with popularity > 80
tempo = spotify["tempo"]
slowest_tempo_index = tempo[is_popular].argmin()
slowest_tempo_song = spotify["track.name"][slowest_tempo_index]
print("The slowest song with popularity > 80 is: " + slowest_tempo_song)
```

### Q2.d

Here is a little code that will create a new column with four (approximately) evenly sized groups based on popularly (low, medium, medium high, and high). The code after creating the columns shows how many rows are in each group. There are a few songs with the same popularity value, which makes it difficult to get exactly evenly sized groups, but this is pretty close. The `(a, b]` output on the left shows the popularity range of that group.

```python
spotify["popgrp"] = pd.qcut(spotify["track.popularity"], 4)
spotify["popgrp"].value_counts()
```

For each group, find the number of explicit songs using the `"track.explicit"` column. Which group has the most explicit songs?

```python
popgrp = spotify.groupby("popgrp")
popgrp["track.explicit"].sum()
# output this to see the groups
print(popgrp["track.explicit"].sum())

print("The group with the most explicit songs is the 4th group, High")
```

### Q2.e

Create a new column that is the product of `danceability` and `energy`. Find the song with the highest product of danceability and energy. Find the tempo of this song. Is it a particularly fast song in terms of tempo? How many songs are faster than this song? How many are slower?

```python
spotify["party"] = spotify["danceability"] * spotify["energy"]

# find the song with the highest party score
max_party_index = spotify["party"].argmax()
max_party_song = spotify["track.name"][max_party_index]
print("The song with the highest party score is: " + max_party_song)

# find the tempo of the song with the highest party score
max_tempo = spotify["tempo"][max_party_index]
print("The tempo of the song with the highest party score is: " + str(max_tempo))

# compare max_tempo with average tempo
mean_tempo = spotify["tempo"].mean()
if max_tempo > mean_tempo:
    print("The song with the highest party score has a tempo higher than the average. Therefore, it is relatively fast.")
else:
    print("The song with the highest party score has a tempo lower than the average. Therefore, it is relatively slow.")

# how many songs are faster than max_tempo
is_faster = spotify["tempo"] > max_tempo
number_of_faster_songs = is_faster.sum()
print("There are " + str(number_of_faster_songs) + " songs that are faster than the song with the highest party score")

# how many songs are slower than max_tempo
is_slower = spotify["tempo"] < max_tempo
number_of_slower_songs = is_slower.sum()
print("There are " + str(number_of_slower_songs) + " songs that are slower than the song with the highest party score")
```
