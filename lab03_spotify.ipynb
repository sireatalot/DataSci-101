### Tuples

Tuples are fixed length lists in Python. Typically, they are small (2 to 5 items). They use round braces to denote:


````python
(3, 2, 1)
````

It's not uncommon to encounter a function that will return a tuple, where each item in the tuple has a given value.


````python
def first_last(w):
  return (w[0], w[-1])

first_last("hello")
````

This is convenient because we can easily assign items in a tuple to variable names using *multiple assignment*:


````python
a, b = first_last("elephant")
print(a)
print(b)
````

### Dictionaries of Lists

For many data analysis problems, we would wish to study a **population**, but instead for reasons of feasibility or possibility, must study a **sample**, a portion of the population.

We think of a sample as composed as **units** (we often write $n$ for the number of units we have). For each unit we measure several things (we will write $k$ to represent the number of things we measure). To organize this data, we create a table with $n$ rows and $k$ columns (a $n \times k$ table).

As we discussed in class, our mental model of tables in Python and Pandas will be "dictionaries of lists (each of the same type, all of the same length)".

## DataFrame and Series

While we can ask a lot of questions using this technique, we can't easily ask questions like, "How many people who got the vaccine got sick?" For this we need more advanced tools.


````python
import pandas as pd

vdf = pd.DataFrame(vaccine)

vdf
````

````python
### pull out just the people who got the vaccine from the "got_sick" column
vdf["got_sick"][vdf["got_vaccine"]].sum()
````

## Grouping

Sometimes, we want to ask questions not just across the entire dataset, but within groups. Dataframes allow us to ask questions like this using the `.groupby` method. Sometimes, you will hear this referred to as a `split-apply-combine` method, where we

1) Split our data by some grouping
2) Apply a function or aggregation to each group
3) Combine the results into a full dataframe

Let's consider an example:

**Which time signature has the highest tempo?**


````python
spotify.groupby("time_signature")["tempo"].max()
````

Let's break this down:

1) We **split** our data by time signature.
2) We applied the **max** function.
3) Python **combined** the mean results for each group into a single, new dataframe. This dataframe is indexed by time signature.

What if we want to do more operations at once? We can use **method chaining**!

Find the Top 5 artists who uses the most keys!


````python
(
  spotify.groupby("artist.name")["key"]
  .size()
  .sort_values(ascending=False)
  .head(5)
)
````

If you put your pandas operations inside parenthesis, you can put each operation on a new line, which dramatically increases the readability of your code.

Let's say we want multiple aggregations. Some common options are supported
by passing strings into the aggregate method.


````python
spotify["duration_s"] = spotify["duration_ms"] / 1000
(
  spotify.groupby("track.explicit")["duration_s"]
  .agg(["count", "max"])
)
````

Are there any artists who don't have any explicit songs?


````python
spotify.groupby("artist.name")["track.explicit"].any()
````
